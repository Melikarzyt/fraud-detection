```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report, f1_score
```

```python
train_data = pd.read_csv('fraud_train.csv')
train_data
```

```python
test_data = pd.read_csv('fraud_test.csv')
test_data
```

```python
df_normal = train_data[train_data['Class'] == 0]
df_fraud = train_data[train_data['Class'] == 1]


Q1 = df_normal.quantile(0.25)
Q3 = df_normal.quantile(0.75)
IQR = Q3 - Q1

condition = ~((df_normal < (Q1 - 1.5 * IQR)) | (df_normal > (Q3 + 1.5 * IQR))).any(axis=1)

df_normal_clean = df_normal[condition]

df_clean = pd.concat([df_normal_clean, df_fraud], axis=0)

print("شکل دیتاست قبل:", train_data.shape)
print("شکل دیتاست بعد:", df_clean.shape)
```

```python
X = df_clean.drop(columns=["Class"])
y = df_clean["Class"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
svc = SVC(kernel='rbf', class_weight='balanced')
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)

f1 = f1_score(y_test, y_pred, average='weighted')
print("Weighted F1:", f1)
print(classification_report(y_test, y_pred))
```

```python
y_test_pred = svc.predict(test_data)

submission = pd.DataFrame({
    "Class": y_test_pred
})

print(submission.shape)
print(submission.head())

submission.to_csv("submission.csv", index=False)
```
