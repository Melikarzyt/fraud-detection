```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report, f1_score
```

```python
train_data = pd.read_csv('fraud_train.csv')
train_data
```

```python
test_data = pd.read_csv('fraud_test.csv')
test_data
```

```python
df_normal = train_data[train_data['Class'] == 0]
df_fraud = train_data[train_data['Class'] == 1]


Q1 = df_normal.quantile(0.25)
Q3 = df_normal.quantile(0.75)
IQR = Q3 - Q1

condition = ~((df_normal < (Q1 - 1.5 * IQR)) | (df_normal > (Q3 + 1.5 * IQR))).any(axis=1)

df_normal_clean = df_normal[condition]

df_clean = pd.concat([df_normal_clean, df_fraud], axis=0)

print("شکل دیتاست قبل:", train_data.shape)
print("شکل دیتاست بعد:", df_clean.shape)
```
